```python
import os
import subprocess
import time

# ---------------------------------------------
# â‘  è‡ªåŠ¨ç”Ÿæˆä¸åŒ freeze_ratio çš„è„šæœ¬
# ---------------------------------------------

freeze_values = [
    0.45,
    0.47,
    0.49,
    0.51,
    0.53,
    0.55,
    0.57,
    0.59,
    0.61,
    0.63,
    0.65
]


# åŸå§‹æ–‡ä»¶è·¯å¾„
src_file = "/data4/huangweigang/gh/csiro-biomass/1114/freeze_ratio_test/DM_pipeline_1114åŸç‰ˆfix.py"

# è‡ªåŠ¨ç»å¯¹è·¯å¾„
src_path = os.path.abspath(src_file)

# ç›®å½•
src_dir = os.path.dirname(src_path)

print(f"åŸæ–‡ä»¶ä½ç½®ï¼š{src_path}")
print(f"ç”Ÿæˆæ–‡ä»¶å°†æ”¾åœ¨ï¼š{src_dir}")

# è¯»å–åŸå§‹æ–‡ä»¶
with open(src_path, "r", encoding="utf-8") as f:
    original_code = f.read()

# ä¿å­˜ç”Ÿæˆçš„è„šæœ¬å
scripts = []

for fr in freeze_values:
    dst_file = f"DM_pipeline_1114åŸç‰ˆfix_fr{int(fr * 100):03d}.py"
    dst_path = os.path.join(src_dir, dst_file)

    # æ›¿æ¢ freeze_ratio è¡Œ
    new_code = original_code.replace(
        '"freeze_ratio"                   : 0.6',
        f'"freeze_ratio"                   : {fr}'
    )

    # å†™å…¥æ–°æ–‡ä»¶
    with open(dst_path, "w", encoding="utf-8") as f:
        f.write(new_code)

    print(f"å·²ç”Ÿæˆï¼š{dst_path}")
    scripts.append(dst_file)

print("å…¨éƒ¨ freeze_ratio è„šæœ¬ç”Ÿæˆå®Œæ¯•ï¼")

```

```python

SCRIPT_DIR = src_dir
PYTHON_BIN = "/home/huangweigang/miniconda3/envs/gh/bin/python"

os.chdir(SCRIPT_DIR)
print(f"\nåˆ‡æ¢åˆ°ç›®å½•ï¼š{SCRIPT_DIR}")
print(f"ä½¿ç”¨ Pythonï¼š{PYTHON_BIN}")

for script in scripts:
    log_file = f"{script}.log"

    print(f"\nåå°è¿è¡Œï¼š{script}")
    print(f"æ—¥å¿—æ–‡ä»¶ï¼š{log_file}")

    run_cmd = f"nohup {PYTHON_BIN} {script} > {log_file} 2>&1 &"

    subprocess.Popen(
        run_cmd,
        shell=True,
        executable="/bin/bash"
    )

    print("ç­‰å¾… 20 ç§’åå¯åŠ¨ä¸‹ä¸€ä¸ªä»»åŠ¡...")
    time.sleep(20)

print("\nğŸ‰ å…¨éƒ¨è®­ç»ƒè„šæœ¬å·²åœ¨åå°å¯åŠ¨å®Œæˆï¼")
print("å¯ä½¿ç”¨ï¼šps -ef | grep DM_pipeline æŸ¥çœ‹è¿è¡ŒçŠ¶æ€")
```
