```python
import random
from pathlib import Path
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt

# 定义扩散模型
def generate_image_with_diffusion(input_image_path, model, device="cuda"):
    # 读取输入图像
    input_img = Image.open(input_image_path).convert("RGB")

    # 使用扩散模型进行生成
    # 你可以为生成过程设置一些额外的条件（如 prompt，噪声级别等）
    generated_img = model(prompt="Generate a new image based on the provided input", init_image=input_img, strength=0.75).images[0]

    return generated_img

# 设置数据目录路径
data_dir = Path(r"D:\DATA_hao\Kaggle_\csiro-biomass\train")

# 获取所有图像文件（.jpg 和 .png）
image_files = list(data_dir.glob("*.jpg")) + list(data_dir.glob("*.png"))

# 检查是否有图像文件
if image_files:
    random_img = random.choice(image_files)  # 随机选择一张图像
    print(f"Selected Image: {random_img.name}")

    # 加载扩散模型
    model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v-1-4-original")
    model.to("cuda")  # 使用GPU加速

    # 生成增强图像
    generated_image = generate_image_with_diffusion(random_img, model)

    # 显示生成的图像
    plt.imshow(generated_image)
    plt.title(f"Generated Image based on {random_img.name}")
    plt.axis("off")  # 不显示坐标轴
    plt.show()

else:
    print("No image files found in the specified directory.")
```

```python
import random
from pathlib import Path
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image
import matplotlib.pyplot as plt

# 定义扩散模型
def generate_image_with_diffusion(input_image_path, model, device="cuda"):
    # 读取输入图像
    input_img = Image.open(input_image_path).convert("RGB")

    # 使用扩散模型进行生成
    # 你可以为生成过程设置一些额外的条件（如 prompt，噪声级别等）
    generated_img = model(prompt="Generate a new image based on the provided input", init_image=input_img, strength=0.75).images[0]

    return generated_img

# 设置数据目录路径
data_dir = Path(r"D:\DATA_hao\Kaggle_\csiro-biomass\train")

# 获取所有图像文件（.jpg 和 .png）
image_files = list(data_dir.glob("*.jpg")) + list(data_dir.glob("*.png"))

# 检查是否有图像文件
if image_files:
    random_img = random.choice(image_files)  # 随机选择一张图像
    print(f"Selected Image: {random_img.name}")

    # 加载扩散模型
    model = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2")

    model.to("cuda")  # 使用GPU加速

    # 生成增强图像
    generated_image = generate_image_with_diffusion(random_img, model)

    # 显示生成的图像
    plt.imshow(generated_image)
    plt.title(f"Generated Image based on {random_img.name}")
    plt.axis("off")  # 不显示坐标轴
    plt.show()

else:
    print("No image files found in the specified directory.")
```
